```python

def html_escape(text):
    return html.escape(text)

def highlight_text(text, all_weight):
    max_alpha = np.max(abs(all_weight))
    highlighted_text = []
    word_index = 0
    for word in text.split():
        weight = all_weight[word_index]
        if weight < 0 :
            highlighted_text.append('<span style="background-color:rgba(135,206,250,' + str(-weight / max_alpha) + ');">' + html_escape(word) + '</span>')
        else:
            highlighted_text.append('<span style="background-color:rgba(250,0,0,' + str(weight / max_alpha) + ');">' + html_escape(word) + '</span>')
        word_index += 1
    highlighted_text = ' '.join(highlighted_text)
    return(highlighted_text)
    
#highlighted_text = highlight_text(decoded, LEG_explanations[0])  
```
